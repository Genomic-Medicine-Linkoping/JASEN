```python
import pandas as pd
```

```python
df = pd.read_csv("summary.tsv", sep="\t", index_col="sample", dtype={'ST':str, 'Type':str})
df
```

```python
validation = pd.read_csv("validation_samples_utf-8.tsv", sep="\t", index_col="sample")
# Change K967 to K967-1
as_list = validation.index.tolist()
idx = as_list.index('K967')
as_list[idx] = 'K967-1'
validation.index = as_list
# Check that the change worked out
validation.loc['K967-1']
```

```python
spaTyper_results = (pd.read_csv("spaTyper.tsv", sep="\t", index_col="sample").drop("species",axis=1)
                    .rename(columns={'Type':'spaType'}))
tseemann_mlst_results = (pd.read_csv("tseemann_mlst.tsv", sep="\t", index_col="sample")
                         .drop("species",axis=1)
                         .rename(columns={'sequence_type':'tseemann_MLST'}))
ariba_mlst_results = (pd.read_csv("ariba_mlst.tsv", sep="\t", index_col="sample")
                         .drop("species",axis=1)
                         .rename(columns={'ST':'ariba_MLST'}))
ariba_local_resistance_DB_results = (pd.read_csv("ariba_local_resistance.tsv", sep="\t", index_col="sample")
                         .drop("species",axis=1)
                         .rename(columns={'ref_name':'ariba_local_resistance_DB'}))
ariba_resfinder_resistance_DB_results = (pd.read_csv("ariba_resfinder_resistance.tsv", sep="\t", index_col="sample")
                         .drop("species",axis=1)
                         .rename(columns={'ref_name':'ariba_resfinder_resistance_DB'}))
aMRFinderPlus_results = (pd.read_csv("aMRFinderPlus.tsv", sep="\t", index_col="sample")
                         .drop("species",axis=1)
                         .rename(columns={'Gene symbol':'aMRFinderPlus'}))
spaTyper_results[spaTyper_results['spaType'].notna()]#.shape[0]
```

```python
ariba_local_resistance_DB = (ariba_local_resistance_DB_results[ariba_local_resistance_DB_results["ariba_local_resistance_DB"].notna()]['ariba_local_resistance_DB']
              .reset_index()
              .drop_duplicates()
              .set_index("sample")
              .groupby(['sample'])['ariba_local_resistance_DB']
              .apply(','.join))
ariba_resfinder_resistance_DB = (ariba_resfinder_resistance_DB_results[ariba_resfinder_resistance_DB_results["ariba_resfinder_resistance_DB"].notna()]['ariba_resfinder_resistance_DB']
              .reset_index()
              .drop_duplicates()
              .set_index("sample")
              .groupby(['sample'])['ariba_resfinder_resistance_DB']
              .apply(','.join))
aMRFinderPlus = (aMRFinderPlus_results[aMRFinderPlus_results["aMRFinderPlus"].notna()]['aMRFinderPlus']
              .reset_index()
              .drop_duplicates()
              .set_index("sample")
              .groupby(['sample'])['aMRFinderPlus']
              .apply(','.join))
ariba_local_resistance_DB
```

Write the merged data to tsv

```python
(validation.merge(right=aMRFinderPlus, how="left", left_index=True, right_index=True)
 .merge(right=spaTyper_results, how="left", left_index=True, right_index=True)
 .merge(right=tseemann_mlst_results, how="left", left_index=True, right_index=True)
 .merge(right=ariba_mlst_results, how="left", left_index=True, right_index=True)
 .merge(right=ariba_local_resistance_DB, how="left", left_index=True, right_index=True)
 .merge(right=ariba_resfinder_resistance_DB, how="left", left_index=True, right_index=True)
 .to_csv("all_data.tsv", sep='\t'))
```
